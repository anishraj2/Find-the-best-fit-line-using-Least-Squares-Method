```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array(eval(input('Enter X values:')))
y=np.array(eval(input('Enter Y values:')))
x_mean=np.mean(x)
y_mean=np.mean(y)
num=0
denum=0
for i in range(len(x)):
    num+=(x[i]-x_mean)*(y[i]-y_mean)
    denum+=(x[i]-x_mean)**2
m=num/denum
print(m)
b=y_mean-m*x_mean
print(b)
y_pred=m*x+b
print(y_pred)
plt.scatter(x,y)
plt.plot(x,y_pred,color='red')
```
